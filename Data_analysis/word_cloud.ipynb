```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#load internet data from CSV file
df = pd.read_csv('it_skills_counts_top.csv')

# generate dictionary of word frequencies
word_freq = dict(zip(df.skill, df['count']))

# generate word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white', max_words=200, max_font_size=200).generate_from_frequencies(word_freq)

# plot the word cloud
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
